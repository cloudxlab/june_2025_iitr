```python
import matplotlib.pyplot as plt

def income_tax(income):
    tax = 0

    slabs = [
        (400000, 0.00),
        (800000, 0.05),
        (1200000, 0.10),
        (1600000, 0.15),
        (2000000, 0.20),
        (2400000, 0.25),
        (float('inf'), 0.30),
    ]

    prev_limit = 0

    for limit, rate in slabs:
        if income > prev_limit:
            taxable = min(income, limit) - prev_limit
            tax += taxable * rate

        if income <= limit:
            break

        prev_limit = limit

    return tax


# Generate datapoints
incomes = list(range(-500000, 1000001, 10000))  # 0 to 50 lakh
taxes = [income_tax(i) for i in incomes]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(incomes, taxes, linewidth=2)

plt.xlabel("Income (₹)")
plt.ylabel("Income Tax (₹)")
plt.title("Indian Income Tax vs Income (New Tax Regime FY 2025-26)")
plt.grid(True)

plt.ticklabel_format(style='plain', axis='both')

plt.show()
```
