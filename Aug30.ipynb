```python
def clean_word(word):
    new_word = []
    for c in word:
        if c.isalpha():
            new_word.append(c)
    return ''.join(new_word)
```

```python
d = {}
for word in words:
    word = clean_word(word.lower())
    key = ''.join(sorted(word))
    if key not in d:
        d[key] = set([word])
    else:
        d[key].add(word)
```

list - [1,2,3]

    append - O(1)
    search: 1 in list: O(n)
    remove or del: O(n)
    sort - n * log (n)
    pop - O(1)
    iterating is allowed
    
    
tuple - (1,2,3,3)

    append - not allowed
    search O(n)
    remove/del/pop - not allowed
    iterating is allowed


dict - {1:1,2:2}
    - key in dict - O(1)
    - d[key] = val - O(1)
    - del d[key] - O(1)
    - takes up too much memory
    
set - {1,2,2}
    - everything about set is same as dictionary just that it doesn't have value.
    - k in set - O(1)
    - add - O(1)
    - remove or delete - O(1)
    - too much memory
