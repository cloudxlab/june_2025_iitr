```python
data_points = np.array([[1,2], [2, 3.9], [3, 6.1]])
x = data_points[:,0]
y = data_points[:, 1]
 # Plot them - scatter

 # Given an equation with m, c find out much error is there in y.
 # y = 4x + 5, at x = 1, 
 # for 1,2 predicted = 9. actual = 2, error = 9 - 2 = 7
 # 

x, y
```

```python
plt.scatter(x, y)
```

```python
def calculate_error(data_points, m, c):
    total_error = 0
    for point in data_points: 
        x, y = point
        y_predict = m*x + c
        error = y_predict - y
        total_error += error*error
    return total_error
```

```python
calculate_error(data_points, 3, 4)
```

```python
calculate_error(data_points, 3, 1)
```

```python
calculate_error(data_points, 2, 1)
```
