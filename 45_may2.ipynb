```python
import numpy as np
import matplotlib.pyplot as plt

# For reproducibility
np.random.seed(42)

# Number of samples
n_samples = 100000

# Generate values uniformly in log10 space: [-5, +5]
log_values = np.random.uniform(-5, 5, n_samples)

log_values = log_values / np.sum(log_values)
# Convert back to original scale: 10^-5 to 10^5
values = 10 ** log_values
values = values / np.sum(values)

# -------------------------------
# Plot original distribution
# -------------------------------
plt.figure(figsize=(8, 5))
plt.hist(values, bins=100)
plt.xscale("log")  # because values span many orders of magnitude
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Random Values (10^-5 to 10^5)")
plt.grid(True)
plt.show()

# -------------------------------
# Take log10 and plot again
# -------------------------------
log_transformed = np.log10(values)

plt.figure(figsize=(8, 5))
plt.hist(log_transformed, bins=100)
plt.xlabel("log10(Value)")
plt.ylabel("Frequency")
plt.title("Distribution After Log Transform")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of samples
n = 10000

# Step 1: Generate uniform random numbers between 0 and 1
x = np.random.uniform(low=0, high=1, size=n)

# Step 2: Plot original distribution
plt.figure(figsize=(8, 5))
plt.hist(x, bins=50)
plt.title("Uniform Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Step 3: Normalize (Z-score normalization)
# mean = 0, std = 1 after normalization
x_normalized = (x - np.mean(x)) / np.std(x)

# Step 4: Plot normalized distribution
plt.figure(figsize=(8, 5))
plt.hist(x_normalized, bins=50)
plt.title("Normalized Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Print statistics
print("Original Mean:", np.mean(x))
print("Original Std :", np.std(x))

print("Normalized Mean:", np.mean(x_normalized))
print("Normalized Std :", np.std(x_normalized))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Training data
x = np.array([0, 1, 3, 4])
y = np.array([9.8, 8.6, 7.0, 6.0])

# Fit linear model (degree 1 polynomial)
m, b = np.polyfit(x, y, 1)

print(f"Slope (m): {m:.4f}")
print(f"Intercept (b): {b:.4f}")

# Predictions
y_pred = m * x + b

# Predict for a new x
x_new = 2
predicted_y = m * x_new + b
print(f"Prediction for x={x_new}: {predicted_y:.4f}")

# Plot original data and fitted line
x_line = np.linspace(min(x), max(x), 100)
y_line = m * x_line + b

plt.scatter(x, y, label="Training Data")
plt.plot(x_line, y_line, label="Linear Model")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```
