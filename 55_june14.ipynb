```python
import gymnasium as gym
import numpy as np
env = gym.make('CartPole-v1')
def play(policy, env, seed):
    
    obs, result = env.reset(seed=seed)
    pos, vel, angle, angvel = obs
    done = False
    count = 0
    while not done:
        pos, vel, angle, angvel = obs
        action = policy(obs)
        obs, reward, done, info, _ = env.step(action)
        count += 1
    return count
```

```python
def test_policy(pl):
    rewards = [play(pl, env, i) for i in range(100)]
    print(np.mean(rewards), np.std(rewards), np.min(rewards), np.max(rewards))
```
