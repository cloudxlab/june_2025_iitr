```python
import numpy as np

x = np.array([171, 173, 163, 162, 160])
y = np.array([1, 1, 0, 0, 0])

def lin_error(m, c, x, y):
    return np.average(((m*x + c) - y)**2)

def sigm(x):
    return np.exp(x)/(np.exp(x) + 1)

def lin_error_sigmoid(m, c, x, y):
    return np.average((sigm(m*x + c) - y)**2)
```

```python
eps = 1e-8
def grad(f, m, c, x, y): # idx = 0, m
    e1 = f(m, c, x, y)
    e1m = f(m + eps, c, x, y)
    e1c = f(m, c + eps, x , y)
    gradm = (e1m - e1)/eps
    gradc = (e1c - e1)/eps
    return gradm, gradc
```

```python
def find_m_c():
    m = 0.12
    c = .3
    eta = .00001
    epoch = 100
    for i in range(epoch):
        gm, gc = grad(lin_error, m, c, x, y)
        m = m - eta * gm
        c = c - eta * gc
        print(i, lin_error(m, c, x, y), f"m:{m}, c:{c}")
```

```python
find_m_c()
```
