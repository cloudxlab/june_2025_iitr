```python
def create_board(sz=8):
    board = []
    for i in range(sz):
        row = []
        board.append(row)
        for j in range(sz):
            row.append(False)
    return board

def print_board(board, sz=8):
    print("-"*(sz*2))
    for row in board:
        line = []
        for v in row:
            if v:
                line.append("x")
            else:
                line.append("_")
        print(' '.join(line))
    print("-"*(sz*2))
    print("\n")
```

```python
def is_free(board, row, col):
    # Check the same col but lowers rows |
    for i in range(row):
        if board[i][col] == True:
            return False
    
    # check the diagonal /
    i, j = row - 1, col - 1
    while i > -1 and j > -1:
        if board[i][j] == True:
            return False
        i -= 1
        j -= 1
    
    # check the diagonal \
    i, j = row - 1, col + 1
    while i > -1 and j < 8:
        if board[i][j] == True:
            return False
        i -= 1
        j += 1
    return True

def find_valid_empty_slots(board, row):
    places = []
    for c in range(8):
        if is_free(board, row, c):
            places.append(c)
    return places
    
def place_queen_in_row(board, row=0):
    if row == 8:
        print_board(board)
    places = find_valid_empty_slots(board, row)
    for col in places: # try each place
        board[row][col] = True
        place_queen_in_row(board, row + 1)
        board[row][col] = False
```

```python
board = create_board()
place_queen_in_row(board, 0)
```
