```python
def f(x):
    return x*x

def dy_dx(f, x):
    eps = 1e-8
    return (f(x+eps) - f(x))/eps
```

```python
dy_dx(f, 10)
```

```python
def loss(o1):
    return (o1-5)**2

dy_dx(loss, 16)
```

```python
def huber(o1):
    a = abs(o1 - 5)
    d = 0.5
    if a <= d:
        # print("Using SQRD")
        return a*a / 2 
    else:
        # print("Using ABS")
        return d * (a - d/2)
```

```python
dy_dx(huber, 10)
```
